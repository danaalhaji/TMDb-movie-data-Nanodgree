```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df_movies = pd.read_csv('tmdb_5000_movies.csv')
df_cred = pd.read_csv('tmdb_5000_credits.csv')
```

***Describe Our Datasets***

```python
df_movies.info()
```

```python
df_cred.info()
```

***Merge Two Datasets***

```python
df_cred.columns = ['id' ,  'title' , 'cast' , 'crew']
df = df_movies.merge(df_cred , on = "id")
```

```python
df.info()
```

```python
df.head()
```

**#drop repeated columns#**

```python
df.drop('title_y', inplace=True, axis=1) 
df.head(1)
```

```python
df.columns
```

```python
df.drop('title_x', inplace=True, axis=1) 
df.head(1)
```

```python
df.columns
```

#rename original_title to title 

```python
df = df.rename(columns={"original_title" : "title"})
```

```python
df.columns
```

***create a copy of our Dataset for any possible mistake so the main dataset will not be affected***

```python
df_copy = df
```

```python
df_copy.info()
```

**Study the Original Language  and Popularity relation**

***1- create dataset***

```python
y = df_copy.groupby('original_language').agg(movie_count=('id', 'size'), mean_popularity=('popularity', 'mean') ).reset_index()
```

```python
y.head()
```

```python
y = y.sort_values(by='movie_count', ascending=False)

fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.bar(y['original_language'],y['movie_count'], color='skyblue', label='Number of Movies')
ax1.set_xlabel('Language')
ax1.set_ylabel('Number of Movies', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticks(range(len(y['original_language'])))
ax1.set_xticklabels(y['original_language'], rotation=45, ha='right')

ax2 = ax1.twinx()
ax2.plot(y['original_language'], y['mean_popularity'], color='salmon', marker='o', label='Mean Popularity')
ax2.set_ylabel('Mean Popularity', color='salmon')
ax2.tick_params(axis='y', labelcolor='salmon')

plt.title('Number of Movies and Mean Popularity by Language')
fig.tight_layout()
plt.show()
```

**Budget Vs Popularity**

```python
y1 = df.groupby(['popularity', 'budget', 'title', 'production_companies','release_date']).agg(budget_mean=('budget', 'mean')).reset_index().sort_values(by='popularity', ascending=False).head(20)
```

```python
y1.columns
```

```python
y1.describe()
```

```python
y1.head()
```

```python
y1.info()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(y1['budget'], y1['popularity'], color='skyblue', edgecolor='black')
plt.title('Budget vs Popularity for Top 20 Most Popular Movies')
plt.xlabel('Budget (10 M)')
plt.ylabel('Popularity')
for i, row in y1.iterrows():
    plt.text(row['budget'], row['popularity'], row['title'], fontsize=8, ha='right')

plt.show()
```

***Compare budget according to year of production and popularity***

```python
df_copy['release_date'] = pd.to_datetime(df['release_date'])
df_copy['year'] = df_copy['release_date'].dt.year
```

```python
df_copy.head()
```

```python
y4 = df_copy.groupby(['year']).agg(mean_budget=('budget', 'mean'),
    sum_movies=('id', 'sum'),
    mean_popularity=('popularity', 'mean')).reset_index()
```

```python
y4.head()
```

```python
y4.describe()
```

```python
y4.info()
```

```python
plt.figure(figsize=(22, 8))
plt.title('Budget vs Year of production')
plt.xlabel('Year Of Production')
plt.ylabel('Budget (10 M)')
plt.bar(y4['year'],y4['mean_budget'] , color ='tab:blue') 
plt.show()
```

```python
# Assuming y4 contains columns: 'year', 'mean_budget', 'mean_popularity'
plt.figure(figsize=(12, 8))

plt.plot(y4['year'], y4['mean_popularity'], color='salmon')
plt.title('Mean Popularity by Year')
plt.xlabel('Year')
plt.ylabel('Mean Popularity')


plt.show()
```
